### **Old Code for Word Similarity**

```python
# Examples of search strings
s1 = 'Bawag bank' # other options: 'Bawag bank', 'Erste', 'Raiffaisen bank'
```

```python
# A naive search method which you need to improve
res = []
for token in banks:
  res.append([s1, token, SequenceMatcher(None, s1, token).ratio()])

df2 = pd.DataFrame(res, columns=['Bank 1', 'Bank 2', 'Score'])
# The outcome is not great, for this search query 'BAWAG Group AG' should have highest similarity
df2.sort_values(by=['Score'], ascending=[False]).head()
```

```python
 #The desired combination has a low score
idx = df2['Bank 2'].isin(['BAWAG Group AG'])

df2[idx].sort_values(by=['Score'], ascending=[False]).head()
```

### **Improved Word Similarity**
Now Testing on Given Data "Bawag Bank

```python
# Query
query = 'Bawag Bank'

# Create TF-IDF vectors for the banks
vectorizer = TfidfVectorizer()
bank_vectors = vectorizer.fit_transform(banks)

# Create a dataframe to store the results
similarity_df = pd.DataFrame(columns=['Bank 1', 'Bank 2', 'Score'])

# Calculate the cosine similarity between the query and each bank
query_vector = vectorizer.transform([query])
for i, bank in enumerate(banks):
    bank_vector = bank_vectors[i]
    similarity = cosine_similarity(query_vector, bank_vector)[0][0]
    similarity_df.loc[i] = [query, bank, similarity]

# Sort the results by similarity score
similarity_df = similarity_df.sort_values(by='Score', ascending=False)

# Print the results
similarity_df
```

## **Results Conclusion**

```python
print("OLD SIMILARITY SCORE")
print(df2[idx].sort_values(by=['Score'], ascending=[False]).head())

print("\nIMPROVED SIMILARITY SCORE")
print(similarity_df[idx].sort_values(by=['Score'], ascending=[False]).head())
```

We can see that we have **improve the similarity score** for the given data **"Bawag Bank"** initially we had the output score of **0.166667** but we improve it and our similarity score reaches to **0.663232**. Which has the **highest similarity score** according to our query.
